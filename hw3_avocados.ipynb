```python
avocado_data['estimated_revenue'] = avocado_data['total_volume'] * avocado_data['averageprice']
avocado_data.head()
```

```python
group_region_type = avocado_data.groupby(['region', 'type'])['averageprice'].mean(numeric_only=True).reset_index()
group_region_type.head()
```

```python
avocado_data.groupby("year")["total_volume"].agg(["mean", "median", "std"]).plot.bar(
    figsize=(12, 6),
    title="Mean, Median, and Standard Deviation of Total Volume by Year",
)
```

```python
bag_totals = avocado_data.groupby('type')[['small_bags', 'large_bags', 'xlarge_bags']].sum()
bag_totals
```
